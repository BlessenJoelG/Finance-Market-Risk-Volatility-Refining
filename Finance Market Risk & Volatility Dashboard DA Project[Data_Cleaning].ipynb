```python
def clean_stock_data(df, ticker, stock_id):

    df['Date'] = pd.to_datetime(df['Date'])

    df['Volume'] = df['Volume'].round().astype('int64')

    df = df.drop_duplicates()

    df['Ticker'] = ticker

    df['Stock_ID'] = stock_id

    return df
```

```python
tata = clean_stock_data(df1,"TATA_STEEL",1)
reliance = clean_stock_data(df2,"RELIANCE",2)
infosys = clean_stock_data(df3,"INFOSYS",3)
icici = clean_stock_data(df4,"ICICI",4)
hdfc = clean_stock_data(df5,"HDFC",5)
```

```python
tata = tata[['Date','Ticker','Close','Stock_ID']].rename(columns={'Close':'Adj_Close'})
reliance = reliance[['Date','Ticker','Close','Stock_ID']].rename(columns={'Close':'Adj_Close'})
infosys = infosys[['Date','Ticker','Close','Stock_ID']].rename(columns={'Close':'Adj_Close'})
icici = icici[['Date','Ticker','Close','Stock_ID']].rename(columns={'Close':'Adj_Close'})
hdfc = hdfc[['Date','Ticker','Close','Stock_ID']].rename(columns={'Close':'Adj_Close'})
```

```python
final_df = pd.concat([tata,reliance,infosys,icici,hdfc])
```

```python
final_df = final_df.sort_values(['Date','Ticker'])

final_df = final_df.reset_index(drop=True)
```
